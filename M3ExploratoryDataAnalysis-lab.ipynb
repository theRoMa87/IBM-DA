## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here
df['ConvertedComp'].plot(kind='box', figsize=(4,3))
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
import numpy as np
Q1,Q3 = df['ConvertedComp'].quantile(.25),df['ConvertedComp'].quantile(.75)
IQR = Q3 - Q1
IQR
```

Find out the upper and lower bounds.


```python
# your code goes here
upperB=Q3+(IQR*1.5)
lowerB=Q1-(IQR*1.5)
print('Upper bound is',upperB,', '
      'Lower Bound is ',lowerB)
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
out=(df['ConvertedComp'] < lowerB) | (df['ConvertedComp'] > upperB)
out.count()
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
df2 = df[(df['ConvertedComp'] > lowerB) & (df['ConvertedComp'] < upperB)]
df2['ConvertedComp'].describe()
```
